```python
%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
#load the UserCarData dataset
car = pd.read_csv('UserCarData.csv')
```

```python
#display the head of the dataset
car.head()
```

```python
#display the bottom 5 rows
car.tail(5)
```

```python
#analytical summary of the dataset
car.describe(include='all')
```

```python
#plot a histogram for all the variables
car.hist(figsize=(20,30))
```

```python
#relationship between categorical and continous variable
sns.boxplot(x="seats", y="engine", data = car)
```

```python
#build a pair plot using the seaborn library
sns.pairplot(car)
```

```python
#Print the total number of duplicate rows.
duplicate_rows_car = car[car.duplicated()]
print("Number of duplicate rows:", duplicate_rows_car.shape)
```

```python
#count the rown before removing the data
car.count()
```

```python
#drop the duplicates
car = car.drop_duplicates()
car.head()
```

```python
#Drop the missing values from the dataset.
#Find the null values
print(car.isnull().sum())
```

```python
#drop the missing values
car = car.dropna()
car.count()
```

```python
#Plot a histogram to find the number of cars per brand.
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make");
```
